## Demo 2: Supplement - Customer Churn Dataset Preprocessing

Data preprocessing of customer service churn dataset (https://www.kaggle.com/blastchar/telco-customer-churn) for classification

```python
# The first part is to import the required packages
import numpy #For array operations
import pandas as pd #DataFrame
import numpy as np

# Load the data (can be found on https://www.kaggle.com/blastchar/telco-customer-churn)
customer_data = pd.read_csv('./data/Telco-Customer-Churn.csv')
customer_data.head()
customer_data = customer_data.replace(r'^\s*$', np.nan, regex=True) #replace empty strings with NaN

# Drop null/error/empty values
# print(customer_data.isnull().sum()) #check null values
customer_data.dropna(inplace=True)
# print(customer_data.isnull().sum()) #check null values again

# Convert TotalCharges to numeric
customer_data['TotalCharges'] = pd.to_numeric(customer_data['TotalCharges'])
customer_data.head()
```

```python
# Changes all values indicating 'no service' to label 'No'

customer_data['MultipleLines']  = customer_data['MultipleLines'].replace({'No phone service' : 'No'})
customer_data['OnlineSecurity']  = customer_data['OnlineSecurity'].replace({'No internet service' : 'No'})
customer_data['OnlineBackup']  = customer_data['OnlineBackup'].replace({'No internet service' : 'No'})
customer_data['DeviceProtection']  = customer_data['DeviceProtection'].replace({'No internet service' : 'No'})
customer_data['TechSupport']  = customer_data['TechSupport'].replace({'No internet service' : 'No'})
customer_data['StreamingTV']  = customer_data['StreamingTV'].replace({'No internet service' : 'No'})
customer_data['StreamingMovies']  = customer_data['StreamingMovies'].replace({'No internet service' : 'No'})
```

```python
# Transform categorical values to dummies
customer_dummies = pd.get_dummies(customer_data[1:], drop_first = True)
customer_dummies.to_csv('./data/Telco-Customer-Churn_dummies.csv')
customer_dummies.head()
```
